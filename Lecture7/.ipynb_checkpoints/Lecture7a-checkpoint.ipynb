# Linear Discriminant Analysis (Simulated Data)

```r
library(MASS)
```

```r
n=200
set.seed(1)

#Generate X data 
Sigma <- matrix(c(1,0,0,1),2,2) #independent covariance matrix
X1 <- mvrnorm(n,c(-1,-1),Sigma) #these are n=200 observations from class 1
X2 <- mvrnorm(n,c(1,1),Sigma) #n=200 observations from class 2
```

Since $n$ is the same for both classes, the prior probability (for any observations) of class membership is the same.

```r
Y1 <- rep(1,n) #generate class labels for "group 1"
Y2 <- rep(2,n) #generate class labels for "group 2"

Y<- c(Y1,Y2) #stack the labels into 1 column vector
X<-rbind2(X1,X2) #stack the Xs into 2 column vectors
sim.data <-data.frame(Y,X) #store in data frame. Note that data from X1 is labelled as "group 1".
```

## Plot the simluated data

```r
plot(X1,col="blue",ylim=c(-4,4),xlim=c(-4,4))
points(X2,col="red")
```

Note that the two classes overlap a bit. LDA will find a linear boudnary between these two distributions to classify them.

```r
lda.fit=lda(Y~X1+X2,data=sim.data)
lda.fit
```

If you multiply each value of LDA1 (the first linear discriminant) by the corresponding elements of the predictor variables and sum them $0.6832674\times X_1 + 0.6736126 \times X_2$ you get a score for each respondent. This score (along with the prior) are used to compute the posterior probability of class membership

```r
plot(lda.fit) # The plots of scores according to the observed class
```

Notice that the score tends to be lower for those observations in group 1.

You can also plot this score directly on the data

```r
lda.fit$scaling[1]
lda.fit$scaling[2]
```

```r
plot(X1,col="blue",ylim=c(-4,4),xlim=c(-4,4))
points(X2,col="red")

lines(-4:4,4:-4,lty="dashed"  ) # the true decision boundary
newX1 <- (-4:4)
newX2 <- (-4:4)

score.coord1 <- lda.fit$scaling[1]*newX1
score.coord2 <- lda.fit$scaling[2]*newX2

lines(score.coord1,score.coord2) # this is the score
```

As the score goes up (>0) the observation will be classified as group 2.

In this simple binary setting (where the priors are equal), we can calculate the (estimated) boundary by averaging the scores across the two groups.

```r
lda.pred <- predict(lda.fit)
(mean(lda.pred$x[Y==2]) + mean(lda.pred$x[Y==1]))/2 # average of mean score for each group
```

Note that this is close to zero.
