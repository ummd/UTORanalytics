# Lecture 6c - The Parameters of the Logistic Function

```r
x <- seq(-4,4, by=0.1)

b0 <- 0 #intercept is 0
b1 <- 1

p <- exp(b0 + b1*x)/(1+exp(b0+b1*x))

plot(x,p)

b0 <- -.5 #intercept is negative
b1 <- 1

p <- exp(b0 + b1*x)/(1+exp(b0+b1*x))

lines(x,p,col='blue')

b0 <- .5 #intercept is positive
b1 <- 1

p <- exp(b0 + b1*x)/(1+exp(b0+b1*x))

lines(x,p,col='red')

b0 <- 0
b1 <- 2 # "slope" is steeper

p <- exp(b0 + b1*x)/(1+exp(b0+b1*x))

lines(x,p,col='green')
```
