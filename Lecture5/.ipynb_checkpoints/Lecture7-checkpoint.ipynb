## Add a non-linear term

The following model specifies a quadratic term for horsepower

```r
lm.fit=lm(mpg~horsepower + I(horsepower^2)) # estimate a linear model regressing sales on TV
summary(lm.fit) # display results
```

### Predict using the fitted values

```r
plot(horsepower,mpg,ylim=c(0,50)) #specify y-axis 

yhat <- predict(lm.fit) # store the fitted values from the model

par(new=TRUE) #create a new plot in same frame, but with different axis
plot(na.omit(horsepower),yhat,type="p",col="blue",ylim=c(0,50),axes=FALSE, ann=FALSE) # also specify y-axis so that second plot lines up with first
# also turn off axis labels and annotations, so that they aren't repeated
```

### Predict using new values

```r
plot(horsepower,mpg)

predicted=data.frame(horsepower=50:230) #Create a new dataframe wtth new values for horsepower (note they are sequential...)
ynew <- predict(lm.fit,predicted) # Get the predicted values from the model for this new dataframe 
predicted$ynew <- ynew # add the predicted values to the new dataframe

lines(predicted$horsepower,predicted$ynew,col="blue") # plot the predicted value (as a line) on the existing axes
```

## Add more non-linear terms!!!
### (The more the better??)

```r
lm.fit=lm(mpg~horsepower + I(horsepower^2) + I(horsepower^3) + I(horsepower^4) + I(horsepower^5)) # estimate a linear model regressing sales on TV
summary(lm.fit) # display results
```

```r
plot(horsepower,mpg)

predicted=data.frame(horsepower=50:230) #Create a new dataframe wtth new values for horsepower (note they are sequential...)
ynew <- predict(lm.fit,predicted) # Get the predicted values from the model for this new dataframe 
predicted$ynew <- ynew # add the predicted values to the new dataframe

lines(predicted$horsepower,predicted$ynew,col="red") # plot the predicted value (as a line) on the existing axes
```

The fit of the model is sooo much better! and look at those siginifincant p-values!

But when adding non-linear terms, we have to be wary of giving the model too much flexibility. This will lead to overfitting and actually decrease prediction performance (see Lecutre 2!)
