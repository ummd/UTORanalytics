### Fit a Cubic Spline

```r
library(splines)

agelims=range(age)
age.grid=seq(from=agelims[1],to=agelims[2]) # construct a grid of values to make predictions at
```

We fit a cubic spline with the function bs()

bs() generates an entire matrix of basis function given the specififed knots.

```r
fit=lm(wage~bs(age,knots=c(25,40,60)),data=Wage) # Uses cubic splines by default

pred=predict(fit,newdata=list(age=age.grid),se=T) #predict from fit at gridpoints
plot(age,wage,col="gray")
lines(age.grid,pred$fit,lwd=2)
lines(age.grid,pred$fit + 2*pred$se,lty="dashed")
lines(age.grid,pred$fit - 2*pred$se,lty="dashed")
```

Produces a spline with six basis functions (3 cubic polynomials + 3 knots), for 7 degrees of freedom (with the intercept)

We can use the df option to produce a spline at uniform quantiles

```r
dim(bs(age,knots=c(25,40,60))) #report degrees of freedom from bs() at 3 knots
```

```r
dim(bs(age,df=6)) #report degrees of freedom from bs() at with 6 degrees of freedom. Nore it is same as above.
```

```r
attr(bs(age,df=6),"knots") #returns the knots (chosen by default at each quartile for 3 knots)
```

### Fit Cubic Spline with knots at quantiles

```r
fit=lm(wage~bs(age,df=6),data=Wage) #let bs() determine knot locations at quantiles

pred=predict(fit,newdata=list(age=age.grid),se=T) #predict at grid
plot(age,wage,col="gray")
lines(age.grid,pred$fit,lwd=2)
lines(age.grid,pred$fit + 2*pred$se,lty="dashed")
lines(age.grid,pred$fit - 2*pred$se,lty="dashed")
```

### Fit a natural spline

```r
fit2=lm(wage~ns(age,df=4),data=Wage) #natural spline with 4 degrees of freedom

pred2=predict(fit2,newdata=list(age=age.grid),se=T)
plot(age,wage,col="gray")
lines(age.grid,pred2$fit,lwd=2,col="red")
lines(age.grid,pred2$fit + 2*pred$se,lty="dashed")
lines(age.grid,pred2$fit - 2*pred$se,lty="dashed")
```

### Fit a Smoothing Spline

```r
plot(age,wage,xlim=agelims,cex=.5,col="darkgray")
title("Smoothing Splines")
fit=smooth.spline(age,wage,df=16)
fit1=smooth.spline(age,wage,cv=TRUE)
fit2$df
lines(fit,lwd=2,col="red")
```

### Fit a Local Regression

```r
plot(age,wage,xlim=agelims,cex=.5,col="darkgray")
title("Local Regression")
fit=loess(wage~age,span=2,data=Wage)
fit2=loess(wage~age,span=.5,data=Wage)
lines(age.grid,predict(fit,data.frame(age=age.grid)),col=c("red"),lty=1,lwd=2,cex=.8)
lines(age.grid,predict(fit2,data.frame(age=age.grid)),col=c("blue"),lty=1,lwd=2,cex=.8)
legend("topright",legend=c("Span=0.2","Span=0.5"),col=c("red","blue"),lty=1, lwd=2, cex=.8)
```

## Fitting GAMs

Ultimating, a GAM is just a big linear regression with a lot of predictors, so we can use lm().

It will automatically implements a natural spline for year and age, and code dummy variables for education

```r
gam1=lm(wage~ns(year,4)+ns(age,5)+education,data=Wage)
```

The gam() library privdes with some tools for plotting a GAM

```r
library(gam)
par(mfrow=c(1,3),pin=c(2,2))
plot.Gam(gam1,se=TRUE,col="red")
```

However, to use a smoothing spline (which cannot be estimated by least squares) we have to use the gam library which contains the function s() for smoothing splines

```r
gam.m3=gam(wage~s(year,4)+s(age,5)+education,data=Wage) #this model includes basis non-linear models of all 3 predictors
```

```r
par(mfrow=c(1,3),pin=c(2,2))
plot(gam.m3,se=TRUE,col="blue")
```

Does including "year" matter?

```r
gam.m1=gam(wage~s(age,5)+education,data=Wage) #model without year
gam.m2=gam(wage~year+s(age,5)+education,data=Wage) #model with year entering linearly
anova(gam.m1,gam.m2,gam.m3,test="F")
```

Conducting an F-Test on the two models suggests that including year is better than not, but a non-linear function of year is not needed.

Therefore gam.m2 is the better model.

We can see this result by looking at the Anove result in the summary() of model M3

```r
summary(gam.m3)
```

The first set of p-values are for including the (parametric) linear term.

The second set are for including the non-linear terms (splines). They correspond to the null of a linear relationship vs. the alterantive of a non-linear relationship.

The large p-value for year suggests that linear is sufficient. However there appears to be a considerable non-linear relationship for age.

We can get predicitons from this model ising the standard predict() function.

```r
preds=predict(gam.m2,newdata=Wage)
```

### GAM with Local Regression

A GAM is _very_ flexible. 

For instance, you can inlcude a local regression of one of the predictors (here with a span of 0.7)

```r
gam.lo=gam(wage~s(year,4)+lo(age,span=0.7)+education,data=Wage)
plot.Gam(gam.lo,se=TRUE,col="green")
```

```r
gam.lo.i=gam(wage~lo(year,age,span=0.5)+education,data=Wage)
```

```r
library(akima)
plot(gam.lo.i)
```

```r
gam.lr=gam(I(wage>250)~year+s(age,df=5)+education,family=binomial,data=Wage)
plot(gam.lr,se=T,col="green")
```
